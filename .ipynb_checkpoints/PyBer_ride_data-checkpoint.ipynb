```python
%matplotlib inline
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Load in csv
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
# Plot the data.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar')
plt.show()
```

```python
import statistics
stdev=statistics.stdev(pyber_ride_df["Avg. Fare ($USD)"])
stdev
```

```python
# Add error bars.
plt.errorbar(pyber_ride_df["Month"], pyber_ride_df["Avg. Fare ($USD)"], yerr=stdev)
plt.show()
```

```python
# Add caps to the error bars.
fig, ax = plt.subplots()
ax.errorbar(pyber_ride_df["Month"], pyber_ride_df["Avg. Fare ($USD)"], yerr=stdev, capsize=3)

plt.show()
```

```python
# Add caps to the error bars
plt.bar(pyber_ride_df["Month"], pyber_ride_df["Avg. Fare ($USD)"], yerr=stdev, capsize=3)
```

```python
# Change the color of the bars to sky blue.
fig, ax = plt.subplots()
plt.bar(pyber_ride_df["Month"], pyber_ride_df["Avg. Fare ($USD)"], color="skyblue", yerr=stdev, capsize=3, label="Avg. Fare ($USD)")
ax.set_yticks(np.arange(0, 51, step=5.0))
# Rotate the labels on the x-axis to horizontal
plt.legend()
plt.xlabel("Month")
# Set the y-axis increment to every $5.
#fig, ax = plt.subplots()
plt.show()
```
